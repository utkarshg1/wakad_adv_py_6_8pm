# Add multiple numbers using function arguments

```python
def add_multiple(*args) -> int | float:
    """
    This function adds multiple numbers from the given numbers
    """
    # Intialize sum to 0
    s = 0

    # Apply for loop on args
    for i in args:
        s = s + i

    # Return the value of sum
    return s
```

```python
b1 = add_multiple(2, 4)
print(b1)
```

```python
b2 = add_multiple(11.5, 12, 13)
print(b2)
```

```python
b3 = add_multiple(3, 4, 5, 11, 12, 13)
print(b3)
```

```python
c = [13, 11, 12, 14, 15, 16, 18, 20, 23, 25]
print(c, type(c))
b4 = add_multiple(*c)
print(b4)
```

```python
d = list(range(1, 1000))
print(len(d))
```

```python
b5 = add_multiple(*d)
print(b5)
```

# Cart billing

```python
cart1 = [
    ("Laptop", 37000, 2),
    ("Mouse", 700, 3),
    ("Keyboard", 1300, 1),
    ("Mouse Pad", 200, 2),
]
```

```python
def billing(*args):
    """This function performs billing for a given cart"""
    # Intialize final total as zero
    final_total = 0

    # Apply for loop on args
    for product, price, qty in args:
        line_total = price * qty
        final_total += line_total  # final_total = final_total + line_total
        print(f"Product name : {product}, Price : {price} INR, Quantity : {qty} Nos")
        print(f"Line Total : {line_total:.2f} INR")
        print("\n" + "=" * 60 + "\n")

    # Print the final amount
    print(f"Final Total of {final_total:.2f} INR")

    # Return the final total
    return final_total
```

# Creating a new user with flexible inputs

```python
def create_user(**kwargs):
    """
    This function is used to create a new user
    """
    # Default user list
    user = {"username": None, "age": None, "email": None}

    # Update the user variable with kwargs
    user.update(kwargs)

    # Return the user
    return user
```

## Using *args and **kwargs

```python
def function3(*args, **kwargs):
    """This function uses both args and kwargs"""
    print(f"Positional arguments : {args}, Type : {type(args)}")
    for i in args:
        print(i)

    print(f"Keyword arguments : {kwargs}, Type : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
function3(3, 4, 5, "Example", True)
```

```python
function3(name="Raman", age=35, salary=12000)
```

```python
function3(12, 13, True, "Example", a=23, b=False, c=45.5, name="Rahul")
```

# Purpose of decorator is to modify behaviour of exisiting function

```python
from typing import Callable, Any
```

```python
def welcome(func: Callable[..., Any]):
    """This function will welcome the user and thank the user"""

    def wrapper(*args, **kwargs):
        print("Welcome user!")
        res = func(*args, **kwargs)
        print(f"Function Results : {res}")
        print("Thank you")
        return res

    return wrapper
```

```python
@welcome
def simple_interest(p: float, n: int, r: float) -> tuple:
    """This function calulates simple interest and amount"""
    i = (p * n * r) / 100
    a = p + i
    return i, a
```

```python
i1, a1 = simple_interest(p=50_000, n=5, r=10.1)
```

```python
a1
```

```python
i2, a2 = simple_interest(25_000, 3, 6.5)
```

```python
import math
```

```python
@welcome
def hypotenuse(a: int | float, b: int | float) -> float:
    """This function calculates hypotenuse for right angled triangle"""
    h = math.sqrt(a**2 + b**2)
    return round(h, 4)
```

# Measure time for any given function

```python
import time
```

```python
time.sleep(1)
print("Hello")
time.sleep(3)
print("World")
```

```python
def measure_time(func: Callable[..., Any]) -> Callable[..., Any]:
    """This decorator is used to measure time for any given function"""

    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        res = func(*args, **kwargs)
        stop = time.perf_counter()
        elapsed = stop - start
        print(f"Function Results : {res}")
        print(f"Time Required : {elapsed:.2f} sec")
        return res

    return wrapper
```

```python
@measure_time
def print_message(message: str, delay: float) -> None:
    time.sleep(delay)
    print(f"The message given : {message}")
```

```python
print_message("Hello Raman!", 2)
```

```python
print_message("Good evening Students!", 3.5)
```

# Get square of numbers add 1 second delay in for loo

```python
@measure_time
def square_list(nums: list) -> list:
    square = []

    for i in nums:
        time.sleep(1)
        s = i**2
        print(f"Square of number {i} is {s}")
        square.append(s)

    return square
```
